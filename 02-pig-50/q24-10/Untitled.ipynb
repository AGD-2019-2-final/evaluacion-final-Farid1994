```python
%load_ext bigdata
%pig_start
%timeout 300
```

```python
%%pig

u = LOAD 'data.csv' USING PigStorage(',') 
    AS (id:int, 
        firstname:CHARARRAY, 
        surname:CHARARRAY, 
        birthday:CHARARRAY, 
        color:CHARARRAY, 
        quantity:INT);


t1= FOREACH u GENERATE TOKENIZE(birthday,'-') AS (m1:bag{t1:tuple(t2:chararray)});
result= FOREACH t1 GENERATE m1.t1.t2;
```

```python
%%pig

DUMP result;
```
